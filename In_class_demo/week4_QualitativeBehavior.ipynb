# In this lab, we study the qualitative behaviors of differential equations by plotting several solutions ($y$ vs. $t$) with differential initial conditions. 

```python
## import packages, and conventional abbrev.
import numpy as np
import scipy as sp
import matplotlib 
import matplotlib.pyplot as plt
```

## We plot a few solutions for $dy/dt = y(y-1)(y-2)$. Recall $dy/dt \approx (y_1 - y_0)/\Delta t$, so $y_1 \approx \Delta t y_0(y_0-1)(y_0-2)+y_0$.

```python
# Define hyper parameters

dt = 0.05
NumStep = 50000
NumSoln = 4
t = [dt * j for j in range(NumStep)]

y0 = [-1, 0.5, 1.5, 3]
```

```python
y = [[], [], [],[]]

for n in range(NumSoln):
    y[n].append(y0[n])

# This is problem 3 of section 2.5 in your textbook. 
def rhs(y):
    return y*(y-1)*(y-2)    

for j in range(1, NumStep):
    for n in range(NumSoln):
        y[n].append(dt * rhs(y[n][j-1]) + y[n][j-1])
```

```python
# Plot the solutions in one figure 
plt.figure(figsize=(10,10))

plt.plot(t,y[0],'b.',markersize=5)

plt.plot(t,y[1],'r.',markersize=5)

plt.plot(t,y[2],'k.',markersize=5)

plt.plot(t,y[3],'g.',markersize=5)

plt.plot(t,0*np.ones(NumStep),'y.',markersize=5)
plt.plot(t,1*np.ones(NumStep),'y.',markersize=5)
plt.plot(t,2*np.ones(NumStep),'y.',markersize=5)

plt.ylim([-10, 10])
plt.xlim([-1, 10])
plt.xlabel('Time')
plt.ylabel('Different solution curves from differential initial conditions')
plt.title('Solutions for $dy/dt = y(y-1)(y-2)$')
```

### You can see from the figure above that 1 is an asymptotically stable equilibrum, and both 0 and 2 are unstable.  

## We plot a few solutions of $dy/dt = y(y-1)^2(y-2)$, where $y=1$ is a semi stable solution. We will see this in the solution graph. 

```python
y00 = [0.9,1.2]
y1 = [[], []]

for n in range(2):
    y1[n].append(y00[n])

# This is problem 7 of section 2.5 in your textbook. 
def rhs1(y):
    return y*(y-1)**2*(y-2)    

for j in range(1, NumStep):
    for n in range(2):
        y1[n].append(dt * rhs1(y1[n][j-1]) + y1[n][j-1])
```

```python
# Plot the solutions in one figure 
plt.figure(figsize=(10,10))

plt.plot(t,y1[0],'b.',markersize=5)

plt.plot(t,y1[1],'r.',markersize=5)

plt.plot(t,1*np.ones(NumStep),'y.',markersize=5)

plt.ylim([0, 2])
plt.xlim([-1, 10])
plt.xlabel('Time')
plt.ylabel('Different solution curves from differential initial conditions')
plt.title('Solutions for $dy/dt = y(y-1)^2(y-2)$')
```
