# In class, we have discussed Problem 25 of Section 2.1:

$ty'+2y = \frac{\sin t}{t},\ t<0$, with initial condition $y(-\frac{\pi}{2})=a$. The solution of the IVP is

$y(t)= \frac{-\cos t+ \frac{\pi^2}{4}a}{t^2}$.

We will look at some solutions of the initial value problem, letting $a$ vary. 

```python
## import packages, and conventional abbrev.
import numpy as np
import scipy as sp
import matplotlib 
import matplotlib.pyplot as plt

## check versions of above packages
print(np.version.version,"\n", sp.__version__,"\n", matplotlib.__version__)
```

```python
size = 5000
end = -1e-1
t = np.linspace(-np.pi/2, end, size)
```

```python
def soln(a):
    return (-np.cos(t) + np.pi**2 / 4 * a) / t ** 2
```

```python
solution = []
ic = np.linspace(-5, 5, 20)

solution = [soln(a) for a in ic]
```

```python
for sol in solution:
    plt.plot(t, sol)

plt.show()
```

## From above, we certainly see a behavior change of the solutions when $a$ varies between -5 and 5, with a critical value somewhere.

## $-\cos t$ is between -1 and 0 on $[-\frac{\pi}{2},0]$, so we test values of $a$ with values greater than, less than, and equal to $\frac{4}{\pi^2}$

```python
values = [4/np.pi**2-1, 4/np.pi**2, 4/np.pi**2+1]
```

```python
solution1 = []

solution1 = [soln(a) for a in values]
```

```python
for sol in solution1:
    plt.plot(t, sol)

plt.show()
```

## We see from above figure that $\frac{4}{\pi^2}$ is the critical value.

```python
for sol in solution1:
    plt.plot(t, sol)

    plt.show()
```

## Now the justification: Consider the behavior of $y(t)= \frac{-\cos t+ \frac{\pi^2}{4}a}{t^2}$ near $t=0$, we look at the Taylor expansion of $-\cos t+ \frac{\pi^2}{4}a=(\frac{\pi^2}{4}a-1)+\frac{t^2}{2!}-\frac{t^4}{4!}+...$

## Making $\frac{\pi^2}{4}a-1=0$ (therefore $a=\frac{4}{\pi^2}$), the solution $y(t)$ looks like $y(t)= \frac{-\cos t+ \frac{\pi^2}{4}a}{t^2}=\frac{\frac{t^2}{2!}-\frac{t^4}{4!}+...}{t^2}=\frac{1}{2}-\frac{t^2}{4!}+\frac{t^4}{6!}-...$. As $t\to0$, the limit will be 0.5.
