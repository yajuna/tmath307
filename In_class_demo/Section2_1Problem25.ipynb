# In class, we have discussed Problem 25 of Section 2.1:

$ty'+2y = \frac{\sin t}{t},\ t<0$, with initial condition $y(-\frac{\pi}{2})=a$. The solution of the IVP is

$y(t)= \frac{-\cos t+ \frac{\pi^2}{4}a}{t^2}$.

We will look at some solutions of the initial value problem, letting $a$ vary. 

```python
## import packages, and conventional abbrev.
import numpy as np
import scipy as sp
import matplotlib 
import matplotlib.pyplot as plt

## check versions of above packages
print(np.version.version,"\n", sp.__version__,"\n", matplotlib.__version__)
```

```python
size = 5000
end = -1e-1
t = np.linspace(-np.pi/2, end, size)
```

```python
def soln(a):
    return (-np.cos(t) + np.pi**2 / 4 * a) / t ** 2
```

```python
solution = []
ic = np.linspace(-5, 5, 20)

solution = [soln(a) for a in ic]
```

```python
for sol in solution:
    plt.plot(t, sol)

plt.show()
```
