# Last week, we have already seen some computational solutions to differential equations together with initial conditions. We will investigate in this demo, more about initial value problems. We will also look at some application problems for first order and their computational solutions.

```python
## import packages, and conventional abbrev.
import numpy as np
import scipy as sp
import matplotlib 
import matplotlib.pyplot as plt
```

### We will look at the same differential equation as last week $dy/dt + 0.5y = 0.5e^{t/3}$ (Example 2 in Section 2.1). The book used IC $y(0)=1$, and we will plot this solution curve for now. 

```python
## Numerical solution with Forward Euler

# define random initial conditions
i_values = np.random.rand(5)
t0 = 0
n = 500 # take more steps to obtain more accurate solutions- more expensive to compute
endTime = 4
t = np.linspace(t0, endTime + t0, n) # 
dt = t[1] - t[0]

def FEsolver(ic,n):
    y = [ic]
    for j in range(1, n):
        y.append(y[j-1] + dt * (-0.5 * y[j-1] + 0.5 * np.exp(t[j-1]/3)))
    return y
    
## Exact solution to compare    

y = []
for ic in i_values:
    y.append(FEsolver(ic,n))
    
## visualize

plt.figure(figsize=(10,10))

plt.plot(t,y[0],'b.',markersize=5)

plt.plot(t,y[1],'r.',markersize=5)

plt.plot(t,y[2],'k.',markersize=5)

plt.plot(t,y[3],'g.',markersize=5)

plt.plot(t,y[4],'c.',markersize=5)

plt.xlabel('Time')
plt.ylabel('Different solution curves from differential initial conditions')
plt.title('Solutions by Forward Euler approximation')#,y=-0.25)
```

### The next example to try is an autonomous equation- these are equations where $dy/dt$ only depends on $y$ explicitly. 

$dy/dt = y(y-2)(y+1)$

with initial conditions $y(0)=-2$, $y(0)=-1/2$, $y(0)=1$, $y(0)=4$.

```python
# define random initial conditions
i_values = [-1.2,-0.5,1,2.1]
t0 = 0.
n = 500 # take more steps to obtain more accurate solutions- more expensive to compute
endTime = .4 # try with 4, 40, 400
t = np.linspace(t0, endTime + t0, n) # 
dt = t[1] - t[0]

def FEsolver(ic,n):
    y = [ic]
    for j in range(1, n):
        y.append(y[j-1] + dt * (y[j-1] * (y[j-1]-2) * (y[j-1]+1)))
    return y
    
## Exact solution to compare    

y = []
for ic in i_values:
    y.append(FEsolver(ic,n))
    
## visualize

plt.figure(figsize=(10,10))

plt.plot(t,y[0],'b.',markersize=5)

plt.plot(t,y[1],'r.',markersize=5)

plt.plot(t,y[2],'k.',markersize=5)

#plt.plot(t,y[3],'g.',markersize=5)

plt.xlabel('Time')
plt.ylabel('Different solution curves from differential initial conditions')
plt.title('Solutions of $dy/dt =y(y-2)(y+1)$')#,y=-0.25)
```
